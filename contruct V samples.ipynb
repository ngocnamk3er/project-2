```python
FILE_PATH = "data/filtered_ReposVul_samples.jsonl"
```

```python
import json

# Đường dẫn đến file .jsonl
file_path = FILE_PATH

samples = []
with open(file_path, 'r', encoding='utf-8') as f:
    for line in f:
        data = json.loads(line)  # Chuyển dòng JSON thành dict
        samples.append(data)

# In ra số lượng mẫu đọc được
print(f"Đã đọc {len(samples)} mẫu.")

# # In thử 5 mẫu đầu tiên để kiểm tra
# for i, sample in enumerate(samples[:5], start=1):
#     print(f"Sample {i}: {sample}")
```

```python
IMPORTANT_KEY = ['index', 'cve_id', 'cwe_id', 'cve_language', 'commit_message', 'code', 'code_before']
```

```python
def flatten_sample(sample, keys):
    result = {}
    for key in keys:
        if key in sample:
            value = sample[key]
            # Nếu là dict → phẳng hóa
            if isinstance(value, dict):
                result.update(value)
            # Nếu là list và phần tử đầu là dict → phẳng hóa phần tử đầu và chỉ lấy code, code_before
            elif isinstance(value, list) and value and isinstance(value[0], dict):
                # Lấy phần tử đầu và chỉ lấy hai trường 'code' và 'code_before'
                result.update({k: v for k, v in value[0].items() if k in ['code']})
            else:
                result[key] = value
    return result



# Tạo danh sách mới đã phẳng hóa
filtered_samples = [flatten_sample(sample, IMPORTANT_KEY) for sample in samples]

# In thử vài kết quả
for i, fs in enumerate(filtered_samples[:3], 1):
    print(f"Filtered Sample {i}: {fs}")
```

```python
import json

# Đường dẫn đến file output
output_path = "data/filtered_ReposVul_V_samples.jsonl"  # Đặt tên file bạn muốn lưu

with open(output_path, 'w', encoding='utf-8') as f:
    for item in filtered_samples:
        json_line = json.dumps(item, ensure_ascii=False)  # Giữ tiếng Việt
        f.write(json_line + '\n')

print(f"Đã lưu {len(filtered_samples)} mẫu vào '{output_path}'.")
```
